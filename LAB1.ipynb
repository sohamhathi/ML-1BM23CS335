```python
import pandas as pd
adult = pd.read_csv("adult.csv")
adult.head()
```

```python
adult.replace("?", pd.NA, inplace=True)
adult.isnull().sum()
```

```python
adult.dropna(inplace=True)
```

```python
categorical_cols = adult.select_dtypes(include=['object']).columns

adult_encoded = pd.get_dummies(adult, columns=categorical_cols, drop_first=True)
```

```python
num_cols = adult_encoded.select_dtypes(include=['int64', 'float64']).columns
```

```python
Q1 = adult_encoded[num_cols].quantile(0.25)
Q3 = adult_encoded[num_cols].quantile(0.75)
IQR = Q3 - Q1
```

```python
adult_encoded = adult_encoded[
    ~((adult_encoded[num_cols] < (Q1 - 1.5 * IQR)) |
      (adult_encoded[num_cols] > (Q3 + 1.5 * IQR))).any(axis=1)
]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
adult_minmax = scaler.fit_transform(adult_encoded)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
adult_standard = scaler.fit_transform(adult_encoded)
```
